```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Diwali_Sales_Data.csv', encoding='latin1')
data
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.drop(['Status','unnamed1'], axis = 1,inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data['Amount'] = data['Amount'].astype('int')
```

```python
data['Amount'].dtypes
```

```python
data.columns
```

```python
data.rename(columns={'Marital_Status':'Wedding'})
```

```python
#Gender
sns.countplot(x='Gender', data=data)
```

```python
ax = sns.countplot(x='Gender', data=data , hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
data.groupby('Gender')['Amount'].sum().plot(kind='bar',figsize=(10,5))
```

```python
#From above graphs we can see that most the buyers are female
```

```python
#Age
sns.countplot(x='Age Group', data=data)
```

```python
x = sns.countplot(x='Age Group', data=data, hue='Gender')
for bars in x.containers:
    x.bar_label(bars)
```

```python
data.groupby('Age Group')['Amount'].sum()
```

```python
data.groupby('Age Group')['Amount'].sum().plot(kind='bar',figsize=(10,5))
```

```python
#from above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#State
sns.countplot(x='State', data=data)
```

```python
plt.figure(figsize=(25,10))
y = sns.countplot(x = 'State', data=data , hue = 'Gender')
for bars in y.containers:
    y.bar_label(bars)

```

```python
data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
data.groupby('State')['Orders'].sum().sort_values(ascending=False).plot(
    kind='bar',
    figsize=(21, 11),
    color='skyblue',
    title='Total Orders by State'
)
```

```python
# from above diagram we can see that order and total amount are from uttar pradesh
```

```python
sns.countplot(x ='Marital_Status' , data=data)
```

```python
z = sns.countplot(x='Marital_Status', data=data, hue='Gender')
for bars in z.containers:
    z.bar_label(bars)
```

```python
data.groupby('Marital_Status')['Amount'].sum().plot(kind='bar',figsize=(10,5))
sns.barplot(data=data,x='Marital_Status', y = 'Amount' , hue = 'Gender')
```

```python
#from above graphs we can see that most buyers are married women
```

```python
#Occupation
a = sns.countplot(x='Occupation', data=data )
for bars in a.containers:
    a.bar_label(bars)
```

```python
data.groupby('Occupation')['Amount'].sum().plot(kind='bar',figsize=(10,5))
```

```python
data.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(10,5))
```

```python
#from above graphs we can see that most of buyers are working in IT , healthcare
```

```python
#product category
plt.figure(figsize=(30,15))
sns.countplot(x='Product_Category', data=data)
plt.show()
```

```python
data.groupby('Product_Category')['Amount'].sum().plot(kind='bar',figsize=(10,5))
```

```python
data.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(10,5))
```

```python
#product id
plt.figure(figsize=(30,15))
sns.countplot(x='Product_ID', data=data)
plt.show()
```

```python
data.groupby('Product_ID')['Orders'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(10,5))
```
